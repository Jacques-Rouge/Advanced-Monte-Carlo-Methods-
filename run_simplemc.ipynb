```python
import os
import importlib
from simplemc_functions import *
```

```python
### My classic Monte Carlo parameters 
N_mc = 20000 # number of Monte Carlo samples
M_mc=5*2**6 ## number of time steps - must be a multiple of 10
T = 1.0

### Payoff parameters
K = 1.0
n_asian = 10 # Number of time steps for Asian option - already defined in Simple_MC_Refactored.py
```

```python

S_T = Run_Simple_MC_S(N_mc, M_mc)
V0_S = np.mean(np.maximum(S_T - K, 0))
print(f"Call Option Price: {V0_S}")

X_t = Run_Simple_MC_X(N_mc, M_mc)
V0_X = np.mean(np.maximum(np.exp(X_t) - K, 0))
print(f"Call Option Price: {V0_X}")
```

```python
S_avg = Run_Simple_MC_Asian_S(N_mc, M_mc, n_asian)
V0_S_asian = np.mean(np.maximum(np.mean(S_avg, axis=1) - K, 0))
print(f"Asian Option Price: {V0_S_asian}")

X_avg = Run_Simple_MC_Asian_X(N_mc, M_mc, n_asian)
V0_X_asian = np.mean(np.maximum(np.mean(np.exp(X_avg), axis=1) - K, 0))
print(f"Asian Option Price: {V0_X_asian}")
```
